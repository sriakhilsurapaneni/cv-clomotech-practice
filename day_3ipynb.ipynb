import numpy as np
import cv2
import matplotlib.pyplot as plt



```python
image=cv2.imread("./img.jpg")
```

```python
!ls
```

```python
def imshow(name,image=None):
  plt.figure(figsize=(10,10))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(name)
  plt.show()
imshow("just a picture",image)
```

```python
h,w=image.shape[:2]
t=np.array([[1,0,h/4],[0,1,w/4]],dtype=np.float32)
x=cv2.warpAffine(image,t,(w,h))
imshow("shifted to right bottom",x)
```

```python
imshow("transpose",cv2.transpose(image))
imshow("transpose",image)
```

```python
t=cv2.getRotationMatrix2D((0,0),45,.5)
imshow("rotated 45",cv2.warpAffine(image,t,(w,h)))
```

```python
imshow('flip',cv2.flip(image,-1))
```
