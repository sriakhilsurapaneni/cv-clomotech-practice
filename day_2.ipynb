```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
def imshow(title,image=None):
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
```

```python
image=np.zeros(shape=(500,500),dtype=np.uint8)

cv2.line(image, (0,0), (500,500), (255,255), 1)
imshow("grey",image)
```

```python
image=np.zeros(shape=(500,500,3),dtype=np.uint8)
cv2.rectangle(image, (100,100), (300,250), (127,50,12), -1)
imshow("rectange",image)
```

```python
image=np.zeros((512,512,3),dtype=np.uint8)
a=np.array([[10,50], [40,50], [90,200], [50,500]])
pts = pts.reshape((-1,1,2))
cv2.polylines(image,[a],True,(12,23,45),8)
imshow("polylines",image)
```

```python
imshow("printing text",cv2.putText(np.zeros((512,512,3),dtype=np.uint8),'Hello World!',(75,290),cv2.FONT_HERSHEY_COMPLEX,1.5,(100,170,0),3))
```
